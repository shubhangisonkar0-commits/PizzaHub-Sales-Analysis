```python
df[(df['IsActiveMember']==1)&(df['Balance']>100000) ] #Active customers with high balance
```

```python
df.sort_values(by='Balance', ascending=False) #sort by balance(highest first)
```

```python
df.rename(columns={'NumOfProducts':'TotalProducts'}, inplace=True) #rename
```

```python
df['AgeGroup']= pd.cut(df['Age'],
                       bins=[18,30,45,60,100],
                       labels=['Young', 'Adult','MidAge','Senior'])
#Age group colunm
```

```python
#balance category
df['BalanceCategory']= np.where(df['Balance']>100000, 'High', 'Low')
```

```python
#covert data type
df['HasCrCard']= df['HasCrCard'].astype('bool')
df['IsActiveMember']= df['IsActiveMember'].astype('bool')
```

```python
df[(df['Exited']==1)&(df['Balance']>10000)] #Customers who left & had high balance
```

```python
df.nlargest(10,'EstimatedSalary') #Top 10 customers by salary
```

```python
df.groupby('Geography')['Balance'].mean() #Average Balance by Geography
```

```python
df.groupby('Gender')['Balance'].mean() #Average Balance by Gender
```

```python
df.groupby('Gender')['Balance'].sum() #Total Balance by Gender
```

```python
df.groupby('Geography')['CreditScore'].agg(['max', 'min']) #Maximum & Minimum Credit Score by Geography
```

```python
df.groupby('Exited')['Age'].mean() #Average Age of Customers Who Exited(1) vs Stayed(0)
```

```python
#balance Statistics
df.groupby('Geography')['Balance'].agg(Total_Balance='sum',
                                       Average_balance='mean',
                                       Max_balance='max',
                                       Min_balance='min',
                                       Count_balance='count'
                                       )
```

```python
df.groupby('TotalProducts')['CustomerId'].count() #Number of Customers by Product Count
```

```python
df.groupby('Geography')['Exited'].mean() #Churn Rate by Geography
```

```python
#1: Geography vs Balance

# Customers from different geographical regions show significant differences in average account balance,
# indicating regional economic variation.
```

```python
#2: High Balance Customers & Churn

#Customers with higher balances tend to exit less frequently, suggesting that high-value customers are more loyal to the bank.
```

```python
#4: Product Usage Pattern

# Most customers own 1–2 bank products, and customers with more products tend to show lower churn,
# implying higher engagement reduces exit probability.
```

```python
#6: Credit Score Distribution

#Customers across all regions have a similar credit score range, indicating credit risk is evenly spread.
```
